## Podpunkt a | Subpoint a

**[PL]**
Odległość (1, 2, 0, 1)T i (0, 0, 0, 1)T

**[EN]**
Distance of (1, 2, 0, 1)T i (0, 0, 0, 1)T


```python
#Customize scipy's hamming distance function
def hamming_distance(a, b):
    if not (isinstance(a, list) and isinstance(b, list)):
        return int(hamming(list(a), list(b)) * len(a))
    return int(hamming(a, b) * len(a))
```

```python
A = Matrix([1, 2, 0, 1])
B = Matrix([0, 0, 0, 1])

print("The distance between vectors:")
print()
pprint([A,B])
print()
print(f"is {hamming_distance(A, B)}.")
```

## Podpunkt b | Subpoint b

**[PL]**
Które są najbliżej w sensie Hamminga?

**[EN]**
Which of the vectors are closest to each other in the Hamming sense?


```python
vectors = [Matrix([1, 2, 1, 2, 0]), Matrix([1, 1, 1, 1, 1]), Matrix([0, 0, 2, 1, 1]), Matrix([2, 2, 2, 1, 0])]
pprint(vectors)
```

```python
def determine_the_minimal_distance(vectors):
    min_distance = len(vectors[0])
    for vector in vectors:
        for another_vector in vectors:
            if vector != another_vector:
                distance = hamming_distance(vector, another_vector)
                if distance < min_distance:
                    min_distance = distance
    return min_distance

def find_the_closest_vector_pairs(vectors):
    min_distance = determine_the_minimal_distance(vectors)
    closest_vector_pairs = []

    for vector in vectors:
        for another_vector in vectors:
            if vector != another_vector:
                distance = hamming_distance(vector, another_vector)
                if distance == min_distance and (vector, another_vector) not in closest_vector_pairs and (another_vector, vector) not in closest_vector_pairs:
                    closest_vector_pairs.append((vector, another_vector))
    print(f"The minimal distance is {min_distance}. The closest vector pairs are:\n\n ")
    return closest_vector_pairs

def vectors_distance_matrix(vectors):
    distance_matrix = []
    for vector in vectors:
        distance_vector = []
        for another_vector in vectors:
            distance = hamming_distance(vector, another_vector)
            distance_vector.append(distance)
        distance_matrix.append(distance_vector)
    return Matrix(distance_matrix)
```

```python
vectors_distance_matrix(vectors)
```

```python
pprint([find_the_closest_vector_pairs(vectors)])
```

## Podpunkt c | Subpoint c

**[PL]**
Znajdź najbardziej odległe od siebie wektory

**[EN]**
Find the most distant vectors from each other


```python
# Find the vector that is furthest from the rest
def find_the_furthest_vector(vectors):
    max_distance = 0
    furthest_vector = None
    for vector in vectors:
        distance = 0
        for another_vector in vectors:
            distance += hamming_distance(vector, another_vector)
        if distance > max_distance:
            max_distance = distance
            furthest_vector = vector
    return furthest_vector

# Find the vector that is closest to the rest
def find_the_closest_vector(vectors):
    min_distance = len(vectors[0]) * len(vectors)
    closest_vector = None

    for vector in vectors:
        distance = 0
        for another_vector in vectors:
            distance += hamming_distance(vector, another_vector)
        if distance < min_distance:
            min_distance = distance
            closest_vector = vector
    return closest_vector
```

```python
print("The furthest vector is:")
pprint(find_the_furthest_vector(vectors))
print()
print("The closest vector is:")
pprint(find_the_closest_vector(vectors))
```
