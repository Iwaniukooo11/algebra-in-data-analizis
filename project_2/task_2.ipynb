**[EN]**
Generate in the selected language all the codewords for the (5, 3)-linear code C over the body Z7 such that the base of the linear code C is  
B = [ (1 0 0 2 4), (0 1 0 1 0), (0 0 1 5 6) ]

```python
import sympy as sp
from sympy import Matrix, pprint
import itertools
```

```python
e1 = Matrix([1, 0, 0, 2, 4])
e2 = Matrix([0, 1, 0, 1, 0])
e3 = Matrix([0, 0, 1, 5, 6])

B = Matrix([list(e1), list(e2), list(e3)]).T
pprint(B)
```

```python
def generate_all_code_words(B: Matrix, modulus: int):
    code_words = []
    for x in itertools.product(range(modulus), repeat = B.shape[1]):
  
        code_word = (B * Matrix(x)) % modulus
        if code_word not in code_words:
            code_words.append(code_word)
    return code_words

pprint(generate_all_code_words(B, 7))
```
